```python
file_path = "VAL-0976_6381-002085207.parquet"
```

```python
from pyspark.sql import SparkSession 

spark = SparkSession.builder \
    .master('local[*]') \
    .appName("Read Project Data") \
    .getOrCreate()
```

```python
from pyspark.sql.types import StructType, StructField, DoubleType, ShortType, FloatType, BinaryType

custom_schema = StructType([
    StructField("x", BinaryType(), True),
    StructField("y", BinaryType(), True),
    StructField("z", BinaryType(), True),
    StructField("intensity", BinaryType(), True),
    StructField("returnnumber", BinaryType(), True),
    StructField("numberofreturns", BinaryType(), True),
    StructField("scandirectionflag", BinaryType(), True),
    StructField("edgeofflightline", BinaryType(), True),
    StructField("classification", BinaryType(), True),
    StructField("synthetic", BinaryType(), True),
    StructField("keypoint", BinaryType(), True),
    StructField("withheld", BinaryType(), True),
    StructField("overlap", BinaryType(), True),
    StructField("userdata", BinaryType(), True),
    StructField("pointsourceid", BinaryType(), True),
    StructField("gpstime", BinaryType(), True),
    StructField("scanchannel", BinaryType(), True),
    StructField("red", BinaryType(), True),
    StructField("green", BinaryType(), True),
    StructField("blue", BinaryType(), True),
    StructField("infrared", BinaryType(), True)
])
```

```python
df = spark.read.format("parquet").schema(custom_schema).option("header", True).load(file_path)
df.printSchema()
```
